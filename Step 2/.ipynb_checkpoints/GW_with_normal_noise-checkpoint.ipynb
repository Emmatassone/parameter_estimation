# Gravitational wave parameter estimation with gaussian noise

```python
from pycbc.waveform import td_approximants, fd_approximants
print("--------------------------------------------------------------")
print("-----List of TIME DOMAIN approximants that are available------")
print("--------------------------------------------------------------")
print(td_approximants())
print("---------------------------------------------------------------")
print("-List of FREQ DOMAIN approximants that are currently available-")
print("---------------------------------------------------------------")
print(fd_approximants())
```

```python
import matplotlib.pyplot as plt
from pycbc import waveform

# Get a time domain waveform, using the SEOBNRv4 EOB model.
# SEOBNRv4 are the two existing semianalytical gravitational waveform models for spin-aligned coalescing black hole
# binaries based on the effective-one-body (EOB) approach. They are informed by numerical relativity 
# simulations and provide the relative dynamics and waveforms from early inspiral to plunge, merger, and ringdown.
# Model paper: https://journals.aps.org/prd/abstract/10.1103/PhysRevD.101.104027
fig,ax = plt.subplots(2)
hp, hc = waveform.get_td_waveform(
                                approximant='SEOBNRv4',
                                mass1 = 6,
                                mass2 = 6,
                                delta_t=1.0/4096,
                                f_lower=40
                                 )
ax[0].plot(hp.sample_times, hp, label='SEOBNRv4')
ax[0].set_ylabel('Strain')
ax[0].set_xlabel('Time (s)')
ax[0].set_xlim((-4,1))
ax[0].legend()

# How to make a ROM model of SEOBNRv4: https://arxiv.org/abs/2008.12932
hp_rom, hc_rom = waveform.get_td_waveform(
                                approximant='SEOBNRv4_ROM',
                                mass1=6,
                                mass2=6,
                                delta_t=1.0/4096,
                                f_lower=40
                                 )
ax[1].plot(hp_rom.sample_times, hp_rom, label='SEOBNRv4_ROM', c='red')
plt.ylabel('Strain')
plt.xlabel('Time (s)')
plt.xlim((-4,1))
plt.legend()
plt.show()
```

## Equal masses Black hole + Gaussian Noise

```python
import numpy as np
t = hp_rom.sample_times
gaussian_noise = np.random.normal(size = t.shape[0])
max_value = np.max(np.abs(np.array(hp_rom)))
scaled_noise = gaussian_noise * max_value
observed_signal = hp_rom + scaled_noise
plt.plot(t, observed_signal, label='observed signal')
plt.legend()
plt.ylabel('Signal')
plt.xlabel('Time (s)')
```
